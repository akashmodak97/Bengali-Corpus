```python
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
import os
cwd = os.getcwd()
print(cwd)
import os.path
completeFolder = os.path.join(cwd,"nltr internship work")

print(completeFolder)
from nltk import RegexpTokenizer
word_tokenize = RegexpTokenizer("[\u0980-\u09FF']+")
#newcorpus = PlaintextCorpusReader(completeFolder,'.*',word_tokenize,encoding="UTF-8") #bengali corpus
newcorpus = PlaintextCorpusReader(completeFolder,'.*') #english corpus
import nltk.corpus
#print(newcorpus.fileids())
print(newcorpus.sents("article35.txt"))
```

```python
from nltk.tokenize import sent_tokenize
A=newcorpus.words("article35.txt")
A
from nltk.probability import FreqDist
fdist = FreqDist()
for word in A:
    fdist[word]+=1
fdist
```

```python
def word_count(list):
    counts = dict()
    words = list
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts
print(word_count(A))
```

```python
import xlsxwriter
workbook = xlsxwriter.Workbook("C:/Users/AKASH/nltr xlsx/Book35.xlsx")
worksheet = workbook.add_worksheet()
dic = word_count(A)
row = 1
col = 1
for key in dic.keys():
    worksheet.write(row,col,key)
    worksheet.write(row,col+1,dic[key])
    row+=1
workbook.close()
```
